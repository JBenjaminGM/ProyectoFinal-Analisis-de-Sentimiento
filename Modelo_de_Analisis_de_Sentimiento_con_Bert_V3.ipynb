Definimos el modelo y el tokenizer

```python
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-multilingual-cased')
model='/content/drive/MyDrive/modeloDefinitivo/'
```

Utilizamos pipeline para hacer predicciones

```python
pipe = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
pipe('La nueva canción del grupo favorito es genial, la repetiré mil veces.')
```

Definimos la funcion que devuelve la interpretación

```python
def get_sentiment(text):
    # Clasificamos la frase
    score = int(pipe(text)[0]['label'])
    # Interpretamos el resultado
    if score == 0:
        sentiment = 'Negativo'
    elif score == 1:
        sentiment = 'Neutral'
    elif score == 2:
        sentiment = 'Positivo'
    else:
        sentiment = 'Valor no válido'
    return sentiment
```

```python
text = 'La nueva canción del grupo favorito es genial, la repetiré mil veces.'
result = pipe(text)[0]['label']
print(result)
print(get_sentiment(text))
```

Cargamos los datos del archivo CSV

```python
import pandas as pd
data = pd.read_csv('datos_limpios.csv', sep=';', encoding='latin1')
data
```

```python
data['Sentimiento'] = data['Tweet'].apply(get_sentiment)
data
```

```python
csv_filename = "data_final.csv"

data.to_csv(csv_filename, index=False, sep=';', encoding='latin-1')

print(f"Se ha creado el archivo CSV '{csv_filename}'.")
```
